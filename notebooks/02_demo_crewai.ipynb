```python
from pathlib import Path

ART = Path("artifacts")
ART.mkdir(exist_ok=True)

item_360.to_parquet(ART / "item_360.parquet", index=False)

# si health es un dataclass/pydantic, lo guardamos como json fácil:
import json
from dataclasses import asdict, is_dataclass

health_dict = asdict(health) if is_dataclass(health) else getattr(health, "model_dump", lambda: health)()
with open(ART / "health_report.json", "w", encoding="utf-8") as f:
    json.dump(health_dict, f, ensure_ascii=False, indent=2)

print("Saved:", ART / "item_360.parquet", "and", ART / "health_report.json")
```
